## Task 2 - Data Exploration

### Function for exploration of numerical columns by providing descriptive statistics and graphical visualisations.

```python
def explore_numerical_column(df, column_name):
    # Descriptive Statistics
    print('Descriptive Statistics:')
    statistics = df[column_name].describe().to_dict()
    statistics_line1 = ", ".join([f"{stat.capitalize()}: {value:.2f}" for stat, value in statistics.items() if stat != 'count'])
    print(statistics_line1)
    
    # Graphical Visualisations 
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Histogram
    hist, bin_edges, _ = ax1.hist(df[column_name], bins=20, edgecolor = 'black', alpha=0.7)
    ax1.set_xlabel(column_name)
    ax1.set_ylabel("Frequency")
    ax1.set_title("Distribution of " + column_name)
    ax1.set_xticks(bin_edges)
    ax1.set_xticklabels([f"{label:.1f}" for label in bin_edges], rotation=90)

    # Boxplot
    ax2.boxplot(df[column_name])
    ax2.set_xlabel(column_name)
    ax2.set_ylabel("Value")
    ax2.set_title("Boxplot of " + column_name)

    plt.tight_layout()
    plt.savefig(column_name + "_exploration.png", dpi=300, bbox_inches="tight")  # Save combined image
    plt.show()

    
```

### Function for exploration of categorical columns by providing descriptive statistics and graphical visualisations.

```python
def explore_categorical_column(df, column_name):
    print('Descriptive Statistics:')
    # Summary
    summary = df[column_name].describe()

    # Unique Categories
    unique_categories = df[column_name].unique()

    # Number of Unique Categories
    num_unique_categories = len(unique_categories)

    # Display Summary and Number of Unique Categories
    print(summary)
    print("Number of Unique Categories:", num_unique_categories)

    # Graphical Visualisation - Bar Plot
    plt.figure(figsize=(10, 8))
    counts = df[column_name].value_counts()
    counts.plot(kind='bar')
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.title("Distribution of " + column_name)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(column_name + "_exploration.png",  dpi=300, bbox_inches="tight")  # Save bar plot image
    plt.show()
```

### 3.2 Training models 

### Function for training and evaluating a machine learning model (KNN or Decision Tree), returning the model, confusion matrices, and evaluation metrics.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

def train_and_evaluate_model(X_train, y_train, X_test, y_test, model_type,cv=5):
    if model_type == "KNN":

        # Set the range of K values to try
        k_values = [3, 5, 7, 10]

        # Perform cross-validation and store the mean accuracy scores
        cv_scores = []
        for k in k_values:
            # Create the KNN classifier
            knn = KNeighborsClassifier(n_neighbors=k)
            
            # Perform cross-validation and compute the mean score
            scores = cross_val_score(knn, X_train, y_train, cv=5)
            
            # Append the mean score to the cv_scores list
            cv_scores.append(scores.mean())

        # Find the best K value
        best_k = k_values[np.argmax(cv_scores)]
        print("Best K:", best_k)
        print("Cross-validation score:",max(cv_scores))
        
        # Train the model using the best K value
        model = KNeighborsClassifier(n_neighbors=best_k)
        
    elif model_type == "DecisionTree":
        max_depth_values = [3, 5, 7, 10]

        cv_scores = []

        # Iterate over each max_depth value
        for max_depth in max_depth_values:
            # Create the Decision Tree Classifier
            dtc = DecisionTreeClassifier(max_depth=max_depth, random_state=42)

            # Perform cross-validation and compute the mean score
            scores = cross_val_score(dtc, X_train, y_train, cv=5)
            mean_score = scores.mean()

            # Append the mean score to the cv_scores list
            cv_scores.append(mean_score)

        # Find the best max_depth value
        best_max_depth = max_depth_values[np.argmax(cv_scores)]
        print("Best max_depth:", best_max_depth)
        print("Cross-validation score:",max(cv_scores))
        
        # Train the model using the best max_depth value
        model = DecisionTreeClassifier(max_depth=best_max_depth)
        

    else:
        raise ValueError("Invalid model_type. Supported values are 'KNN' and 'DecisionTree'.")
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on training and test sets
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    #Create confusion matric for both training and test set
    train_confusion_matrix = confusion_matrix(y_train, y_train_pred)
    test_confusion_matrix = confusion_matrix(y_test, y_test_pred)
    
    # Evaluate performance metrics
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    train_precision = precision_score(y_train, y_train_pred, average='weighted', zero_division=0, labels=np.unique(y_train_pred))
    test_precision = precision_score(y_test, y_test_pred, average='weighted', zero_division=0, labels=np.unique(y_test_pred))
    train_recall = recall_score(y_train, y_train_pred, average='weighted', zero_division=0, labels=np.unique(y_train_pred))
    test_recall = recall_score(y_test, y_test_pred, average='weighted', zero_division=0, labels=np.unique(y_test_pred))
    train_f1 = f1_score(y_train, y_train_pred, average='weighted', zero_division=0, labels=np.unique(y_train_pred))
    test_f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=0, labels=np.unique(y_test_pred))
    
    # Store the evaluation metrics in a dictionary
    
    evaluation_metrics = {
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Train Precision': train_precision,
        'Test Precision': test_precision,
        'Train Recall': train_recall,
        'Test Recall': test_recall,
        'Train F1 Score': train_f1,
        'Test F1 Score': test_f1
    }
    
    return model, train_confusion_matrix, test_confusion_matrix, evaluation_metrics
```

### Function to print confusion matrix

```python
def print_confusion_matrix(model, suite, train_cm, test_cm, labels):
    # Create subplots for KNN and DT confusion matrices
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))

    # Plot Training set confusion matrix heatmap
    axs[0].imshow(train_cm, cmap='Blues')
    axs[0].set_xticks(np.arange(len(labels)))
    axs[0].set_yticks(np.arange(len(labels)))
    axs[0].set_xticklabels(labels, rotation=90)
    axs[0].set_yticklabels(labels)
    axs[0].set_title(model + ' Confusion Matrix for Training Set')
    axs[0].set_xlabel('Predicted label')
    axs[0].set_ylabel('True label')

    # Plot Test Set confusion matrix heatmap
    axs[1].imshow(test_cm, cmap='Blues')
    axs[1].set_xticks(np.arange(len(labels)))
    axs[1].set_yticks(np.arange(len(labels)))
    axs[1].set_xticklabels(labels, rotation=90)
    axs[1].set_yticklabels(labels)
    axs[1].set_title(model + ' Confusion Matrix for Test Set')
    axs[1].set_xlabel('Predicted label')
    axs[1].set_ylabel('True label')

    # Loop over data dimensions and create text annotations
    for i in range(len(labels)):
        for j in range(len(labels)):
            axs[0].text(j, i, train_cm[i, j], ha="center", va="center", color="black")
            axs[1].text(j, i, test_cm[i, j], ha="center", va="center", color="black")

    plt.tight_layout()
    plt.savefig(model + "_" + suite + "_confusion_matrix.png", dpi=300, bbox_inches="tight")

    plt.show()
```

### Bar plots comparing the performance of two models. Each bar plot represents the accuracy scores of the models on both the training and test sets across different suites

```python
# Set the width of each bar
bar_width = 0.35

# Define the index for the x-axis
index = np.arange(len(metrics_df['Suite']))

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot for Decision Tree model
axes[0].bar(index - bar_width/2, metrics_df['DT Train Accuracy'], width=bar_width, color='red', label='Train Accuracy')
axes[0].bar(index + bar_width/2, metrics_df['DT Test Accuracy'], width=bar_width, color='blue', label='Test Accuracy')
axes[0].set_xlabel('Suite')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Decision Tree Model')
axes[0].set_xticks(index)
axes[0].set_xticklabels(metrics_df['Suite'])
axes[0].legend(loc='lower right')

# Add labels to the bars in the Decision Tree plot
for i, v in enumerate(metrics_df['DT Train Accuracy']):
    axes[0].text(i - bar_width/2, v + 0.01, str(round(v, 2)), ha='center', color='black')
for i, v in enumerate(metrics_df['DT Test Accuracy']):
    axes[0].text(i + bar_width/2, v + 0.01, str(round(v, 2)), ha='center', color='black')

# Plot for KNN model
axes[1].bar(index - bar_width/2, metrics_df['KNN Train Accuracy'], width=bar_width, color='red', label='Train Accuracy')
axes[1].bar(index + bar_width/2, metrics_df['KNN Test Accuracy'], width=bar_width, color='blue', label='Test Accuracy')
axes[1].set_xlabel('Suite')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('KNN Model')
axes[1].set_xticks(index)
axes[1].set_xticklabels(metrics_df['Suite'])
axes[1].legend(loc='lower right')

# Add labels to the bars in the KNN plot
for i, v in enumerate(metrics_df['KNN Train Accuracy']):
    axes[1].text(i - bar_width/2, v + 0.01, str(round(v, 2)), ha='center', color='black')
for i, v in enumerate(metrics_df['KNN Test Accuracy']):
    axes[1].text(i + bar_width/2, v + 0.01, str(round(v, 2)), ha='center', color='black')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

plt.savefig("comparison_bar_plot",  dpi=300, bbox_inches="tight")
plt.show()
```

### Line Plot to compare the train and test precision, recall, and F1 scores of the KNN and DT models

```python
# Extract the required data from the summary table
suites = metrics_df['Suite']
knn_train_precision = metrics_df['KNN Train Precision']
knn_test_precision = metrics_df['KNN Test Precision']
knn_train_recall = metrics_df['KNN Train Recall']
knn_test_recall = metrics_df['KNN Test Recall']
knn_train_f1 = metrics_df['KNN Train F1 Score']
knn_test_f1 = metrics_df['KNN Test F1 Score']
dt_train_precision = metrics_df['DT Train Precision']
dt_test_precision = metrics_df['DT Test Precision']
dt_train_recall = metrics_df['DT Train Recall']
dt_test_recall = metrics_df['DT Test Recall']
dt_train_f1 = metrics_df['DT Train F1 Score']
dt_test_f1 = metrics_df['DT Test F1 Score']

# Set the figure size
plt.figure(figsize=(10, 6))

# Line plot for KNN precision, recall, and F1 score
plt.plot(suites, knn_train_precision, marker='o', label='KNN Train Precision')
plt.plot(suites, knn_test_precision, marker='o', label='KNN Test Precision')
plt.plot(suites, knn_train_recall, marker='o', label='KNN Train Recall')
plt.plot(suites, knn_test_recall, marker='o', label='KNN Test Recall')
plt.plot(suites, knn_train_f1, marker='o', label='KNN Train F1 Score')
plt.plot(suites, knn_test_f1, marker='o', label='KNN Test F1 Score')

# Line plot for DT precision, recall, and F1 score
plt.plot(suites, dt_train_precision, marker='o', label='DT Train Precision')
plt.plot(suites, dt_test_precision, marker='o', label='DT Test Precision')
plt.plot(suites, dt_train_recall, marker='o', label='DT Train Recall')
plt.plot(suites, dt_test_recall, marker='o', label='DT Test Recall')
plt.plot(suites, dt_train_f1, marker='o', label='DT Train F1 Score')
plt.plot(suites, dt_test_f1, marker='o', label='DT Test F1 Score')

# Set the labels, title, and legend
plt.xlabel('Suites')
plt.ylabel('Score')
plt.title('Comparison of Train and Test Precision, Recall, and F1 Scores: KNN vs DT')
plt.legend()
plt.xticks(rotation=45)

plt.savefig("comparison_line_plot",  dpi=300, bbox_inches="tight")
plt.show()
```
